```python
import requests
from bs4 import BeautifulSoup

def scrape_von_yor_by_keywords(start, end, keywords):
    data = []
    headers = {'User-Agent': 'Mozilla/5.0'}  # Basic user-agent header

    for keyword in keywords:
        for pageNo in range(start, end + 1):
            url = f"https://von.gov.ng/?s={keyword}&page={pageNo}"
            html_text = requests.get(url, headers=headers, verify=False).text
            soup = BeautifulSoup(html_text, 'html.parser')
            articles = soup.find_all('h2', class_="title")

            for article in articles:
                title = article.a.text.strip()
                link = article.a["href"]
                _article_page = requests.get(link, headers=headers, verify=False).text
                _page_soup = BeautifulSoup(_article_page, 'html.parser')

                # Check if the image meta tag exists and extract content
                meta_image = _page_soup.find('meta', property='og:image')
                image_link = meta_image['content'] if meta_image else "No image available"

                _article_div = _page_soup.find('div', class_='entry-content clearfix single-post-content')

                # Extract text from the article or provide a default message
                if _article_div:
                    article_text = '\n'.join(p.get_text(strip=True) for p in _article_div.find_all('p') if p.get_text(strip=True))
                else:
                    article_text = "No content available"

                # Extract publication time or provide a default message
                time_tag = _page_soup.find('time', class_='post-published updated')
                published_time = time_tag['datetime'] if time_tag else "No published time available"

                # Store all extracted data in a dictionary and append to the data list
                data.append({
                    'Keyword': keyword,
                    'Published_time': published_time,
                    'Title': title,
                    'Full_Content': article_text,
                    'Article URL': link,
                    'Image_link': image_link
                })

    return data
```

```python
keywords = ['technology', 'health', 'education']  # Define the three keywords you want to search for
scraped_articles = scrape_von_yor_by_keywords(1, 2, keywords)  # Scrape from page 1 to 2 for each keyword
print(scraped_articles)
```

```python
with open('scraped_articles.txt', 'w') as file:
    for article in scraped_articles:

        values = [str(value) for value in article.values()]

        article_str = '\n'.join(values)
        file.write(article_str + "\n\n")

print("Articles have been saved to 'scraped_articles.txt'.")
```
